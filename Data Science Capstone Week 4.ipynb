## Introduction:
Stafford, New Jersey is a quiet town on the coast where many people from the surrounding metropolitan areas of New York and Philadelphia come to retire.  There are many medical facilities in close proximity to this area that people in the retiree demograhpic need to visit on a regular basis.  Many retirees are not able to attend these appointments because of transporation issues.  I believe a shuttle service to transport people who are unable to drive to and from their medical facilty would be a valuable service to this commutnity.  There is an ample network of medical facilites in close proximity to one another that would make a pickup/dropoff service feasable. 

## Data:
Medical facility data within proximity to retiree housing and community centers will gathered to determine the best pickup and dropoff locations for the shuttle to serve the maximum amount of people.   It will also be crucial in determining the most effcient routes of travel.   This data will be sourced from FourSquare API and Census.gov to identify where medical facilites are in relation to Stafford township and also to identify the demographic of specific areas within the town.
